## This notebook was used to create single-year GDP files for QGIS

```python
import pandas as pd
```

```python
indicators1 = pd.read_csv("indicators1998-2003.csv", sep=';', decimal=',')
indicators2 = pd.read_csv("indicators2006-2011.csv", sep=';', decimal=',')

#drop duplicate entries
indicators1 = indicators1.drop_duplicates(subset=["country", "year"])
indicators2 = indicators2.drop_duplicates(subset=["country", "year"])
```

```python
# checking for consistent entries
print(indicators1.groupby("year")["country"].count())
print(indicators2.groupby("year")["country"].count())
```

## 

## Create one file per year, containing columns: year / country / GDP 

### Slice 1

```python
indicators1["GDP"] = (indicators1["GDP"].astype(str).str.replace(",", ".", regex=False) #fixing reading issues with comma/periods
                      .pipe(pd.to_numeric, errors="coerce")
)
for year in range(1998, 2004):
    df = indicators1[indicators1["year"] == year][["year", "country", "GDP"]]
    df.columns = ["year", "eez_country", "GDP"]
    #df.to_csv(f"gdp_{year}.csv", index=False) #uncomment to create file
    print(f"gdp_{year}.csv — {len(df)} rows")
    #print(df) #uncomment to see df
```

### Slice 2

```python
indicators2["GDP"] = (indicators2["GDP"].astype(str).str.replace(",", ".", regex=False)#fixing reading issues with comma/periods
    .pipe(pd.to_numeric, errors="coerce")
)
 
for year in range(2006, 2012):
    df = indicators2[indicators2["year"] == year][["year", "country", "GDP"]]
    df.columns = ["year", "eez_country", "GDP"]
    #df.to_csv(f"gdp_{year}.csv", index=False) #uncomment to create file
    print(f"gdp_{year}.csv — {len(df)} rows")
    #print(df) #uncomment to see df
```
