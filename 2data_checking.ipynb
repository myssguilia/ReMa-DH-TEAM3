```python
import matplotlib
import pandas as pd
from scipy import stats
import numpy as np
```

```python
pirates1 = pd.read_csv("pirates_slice1.csv", sep=";", decimal=',')
pirates2 = pd.read_csv("pirates_slice2.csv", sep=";", decimal=',')

indicators1 = pd.read_csv("indicators1998-2003.csv", sep=";", decimal=',')
indicators2 = pd.read_csv("indicators2006-2011.csv",sep=";", decimal=',' )

print(pirates1.shape)
print(pirates2.shape)
```

## Checking for missing EEZ entries, slice 1

```python
missing_eez1 = pirates1[pirates1["eez_country"].isna()]

print(f"Rows with missing eez_country: {len(missing_eez1)}")
print(f"Out of total: {len(pirates1)}")
print()
print(missing_eez1[["date", "attack_type", "location_description", "nearest_country", "eez_country",
                     "latitude", "longitude", "shore_distance", "shore_longitude", "shore_latitude"]].to_string())
```

## Checking for missing EEZ entries, slice 2

```python
missing_eez2 = pirates2[pirates2["eez_country"].isna()]
 
print(f"Rows with missing eez_country: {len(missing_eez2)}")
print(f"Out of total: {len(pirates2)}")
print()
print(missing_eez2[["date", "attack_type", "location_description", "nearest_country", "eez_country",
                     "latitude", "longitude", "shore_distance", "shore_longitude", "shore_latitude"]].to_string())
```

## Check for missing attack type, slice 1

```python
missing_attack1 = pirates1[pirates1["attack_type"].isna()]
 
print(f"Rows with missing attack_type: {len(missing_attack1)}")
print(f"Out of total: {len(pirates1)}")
print()
print(missing_attack1[["date", "attack_type", "location_description", "nearest_country", "eez_country"]].to_string())
 
```

## Check for missing attacks, slice 2

```python
missing_attack2 = pirates2[pirates2["attack_type"].isna()]
 
print(f"Rows with missing attack_type: {len(missing_attack2)}")
print(f"Out of total: {len(pirates2)}")
print()
print(missing_attack2[["date", "attack_type", "location_description", "nearest_country", "eez_country"]].to_string())
```

## Check for missing GDP in general, slice 1

```python
missing_gdp1 = indicators1[indicators1["GDP"].isna()]
 
print(f"Rows with missing GDP: {len(missing_gdp1)}")
print(f"Out of total: {len(indicators1)}")
print()
print("Missing by country:")
print(missing_gdp1.groupby("country")["year"].apply(list).to_string())
```

## Check for missing GDP in countries involved in attacks, slice 1 (EMPTY BECAUSE THE FILE WAS ENRICHED DURING PREPROCESSING)

```python
unique1= pirates1['eez_country'].unique()
unique2 = pirates2['eez_country'].unique()
#print(unique2)
```

```python
missing_in_attacks = missing_gdp1[missing_gdp1["country"].isin(unique1)]
 
print(f"Attack countries with missing GDP: {len(missing_in_attacks)}")
print()
print("Missing by country and year:")
print(missing_in_attacks.groupby("country")["year"].apply(list).to_string())
```

## Check for missing GDP in general, slice 2

```python
missing_gdp2 = indicators2[indicators2["GDP"].isna()]
 
print(f"Rows with missing GDP: {len(missing_gdp2)}")
print(f"Out of total: {len(indicators2)}")
print(f"Percentage: {len(missing_gdp2)/len(indicators2)*100:.1f}%")
print()
print("Missing by country:")
print(missing_gdp2.groupby("country")["year"].apply(list).to_string())
```
